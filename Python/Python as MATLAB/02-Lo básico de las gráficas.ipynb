## 1. <span id="introduccion">Introducción</span>
Como se mostró en "Comienzos.ipynb", es necesario para tener una respuesta gráfica, el uso de una librería de Python. Se mostrará, únicamente, los comandos básicos equivalentes a los que figuran en la guía "Introduccion_a_Matlab_y_a_Simulink_para_Control.pdf".
Existen dos formas de adentrarse en el uso de dicha librería. Una orientada a objetos, y otra utilizando la interfaz pyplot. En este caso, utilizaremos el segundo camino.

```python
import matplotlib.pyplot as plt
import numpy as np
import control.matlab as ctrl
```

## 2. Plot
Para graficar cualquier función en general utilizaremos el comando plot, que sólo necesita definir vector de abscisas y vector de ordenadas

```python
# Ejemplo función cuadrática
# Definimos el vector de abscisas
x = np.arange(-5, 5, .1)   # una alternativa es np.linspace(start, stop, size)

# Definimos el vector de ordenadas
y = x ** 2

# Realizamos la gráfica
plt.plot(x, y)
```

Si imprimimos por pantalla cada vector, veríamos que cada elemento de ordenada corresponde a cada elemento de abscisa:

```python
# Imprimimos ambos vectores
# notese que ambos vectores tienen dimensiones iguales
print("Abscisas:", x)
print("Ordenadas:", y)
```

### 2.A. <span id="que-sucede-solo-un-vector">¿Qué sucede si solo utilizamos un vector?</span>
Veremos a continuación la respuesta a utilizar sólo el vector de ordenadas con plot:

```python
plt.plot(y)
```

Si observamos bien, veremos que el intervalo del eje de abscisas ahora es de [0 , 100], pero la gráfica es la misma. Se obtiene un gráfico donde el eje de abscisas será la posición del elemento del vector y la ordenada el valor que tiene el vector en dicha posición.

### 2.B. Etiquetas y anotaciones
Recurrimos a las etiquetas y anotaciones para agregar información al gráfico

#### Etiquetas
Utilizamos `plt.xlabel('x')` y `plt.ylabel('y')` para establecer las etiquetas del eje 'x' y del eje 'y'

```python
plt.plot(x, y)
plt.xlabel('Eje x')
plt.ylabel('Eje y')
```

La función `plt.title('Quadratic Function')` es usada como título del gráfico

```python
plt.plot(x, y)
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Función Cuadrática')
```

## 3. Adicionales
### 3.A. Grilla
Podemos agregar una grilla para facilitar la lectrura del gráfico con la función`plt.grid(True)`

```python
plt.plot(x, y)
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Función Cuadrática')
plt.grid(True)
```

### 3.B. Show
Utilizaremos la función `plt.show()` para mostrar el gráfico en una nueva ventana si estamos utilizando los comandos desde una terminal. En el caso de Jupyter Notebook, las celdas son ejecutadas por el kernel y el gráfico es mostrado "en línea". De todas formas, se agregará dicha línea en el caso de querer copiar el código en un script o terminal.

```python
plt.plot(x, y)
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Función Cuadrática')
plt.grid(True)
plt.show()
```

Existe la posibilidad, en el caso de querer utilizar más de una figura (vease más adelante), y estemos desde una terminal, veremos que el comando `show()` es bloqueante, es decir, no se mostrará la siguiente figura hasta que la actual no sea cerrada. Para que se entienda mejor, se da a continuación dos ejemplos para compilar desde un script de Python o desde su intérprete:

```python
# Misma función que venimos tratando
x = np.arange(-5, 5, .1)
y = x ** 2

# primera figura que se abrirá en una ventana
plt.figure(1)
plt.plot(x, y, '-.rx')
plt.show()
# segunda figura que se abrirá cuando se cierre la figura anterior
plt.figure(2)
plt.plot(y, x, '--mo')

# pasamos un argumento para que la figura 3 se abra junto con la 2
plt.show(block=False)

plt.figure(3)
plt.plot(x, y, ':bs')

#se recomienda dejar el último show sin el parámetro block
#para evitar que el programa finalice
plt.show()
```

## 4. <span id="multiples-graficas-y-leyenda">Múltiples gráficas y leyenda</span>
### 4.A. <span id="unica-figura">Única figura</span>
Supongamos que queremos graficar tres sinusoides con frecuencias diferentes, ingresemos en un archivo nuevo:

```python
t = np.arange(0, 2*np.pi, np.pi/20)
y1 = np.sin(t)
y2 = np.sin(t-np.pi/2)
y3 = np.sin(t-np.pi)

plt.plot(t,y1)
plt.plot(t,y2)
plt.plot(t,y3)
plt.xlabel('t')
plt.ylabel('y')
plt.title('Sine Waves')
plt.show()
```

En este caso, la librería asigna colores a las diferentes funciones, aunque podemos manualmente modificar dichos colores, su estilo y además añadir una leyenda.

| Espec. | Color              |
|--------|--------------------|
| r      | Rojo               |
| b      | Azul (por defecto) |
| w      | Blanco             |
| g      | Verde              |
| c      | Cian               |
| m      | Magenta            |
| y      | Amarillo           |
| k      | Negro              |


| Espec. | Estilo de línea            |
|--------|----------------------------|
| -      | Línea sólida (por defecto) |
| --     | Línea rayada               |
| :      | Línea punteada             |
| -.     | Línea punto-raya           |


| Espec. | Estilo de marca        |
|--------|------------------------|
| +      | Signo más              |
| o      | Círculo                |
| .      | Punto                  |
| *      | Asterisco              |
| s      | Cuadrado               |
| d      | Diamante               |
| x      | Cruz                   |
| p      | Estrella de 5 puntas   |
| h      | Estrella de 6 puntas   |


```python
plt.plot(t, y1, '-.rx')
plt.plot(t, y2, '--mo')
plt.plot(t, y3, ':bs')

plt.xlabel('t')
plt.ylabel('y')
plt.title('Sine Waves')
plt.legend(['y1', 'y2', 'y3'])
plt.grid(True)
plt.show()
```

### 4.B. <span id="multiples-figuras">Múltiples figuras</span>
También podrı́amos querer cada gráfico en una figura diferente. Para ello debemos ejecutar el comando `figure(2)` antes de graficar por segunda vez y `figure(3)` antes del último gráfico.

```python
plt.plot(t, y1, '-.rx')
plt.figure(2)
plt.plot(t, y2, '--mo')
plt.figure(3)
plt.plot(t, y3, ':bs')
```

### 4.C. Subplot
Otra opción válida, es que las tres funciones aparezcan en una sola figura pero las tres graficadas en forma independiente. Para ello utilicemos `subplot(m,n,p)`, que dividirá a la figura en *m* filas y *n* columnas, pero crea una figura en la posición *p*.

```python
plt.subplot(1, 3, 1)  # dividimos la figura en 1 fila y 3 columnas
                      # esta grafica irá en la primera columna
plt.plot(t, y1, '-.rx')
plt.subplot(1, 3, 2)  # esta grafica irá en la segunda columna
plt.plot(t, y2, '--mo')
plt.subplot(1, 3, 3)  # esta grafica irá en la tercera columna
plt.plot(t, y3, ':bs')
plt.show()
```
